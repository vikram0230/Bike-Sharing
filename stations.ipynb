```python
import json

# Assuming the JSON data is in a file named 'stations.json'
with open('stations.json', 'r') as file:
    json_data = json.load(file)

# Display the JSON data
json_data['data']['stations']
```

```python
len(json_data['data']['stations'])
```

```python
import pandas as pd

stations_df = pd.DataFrame(json_data['data']['stations'])
stations_df.drop(columns=['rental_uris','region_id','address','short_name'], inplace=True)
# stations_df.index = stations_df['station_id']
stations_df.to_csv('stations.csv', index=False)
```
